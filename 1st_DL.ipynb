```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist

# Load the dataset
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# Class names corresponding to labels
class_names = ["T-shirt", "Trouser", "Pullover", "Dress", "Coat",
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

# Dictionary to store one sample per class
sample_images = {}
for img, label in zip(train_images, train_labels):
    if label not in sample_images:
        sample_images[label] = img
    if len(sample_images) == 10:  # Stop once we have all 10 classes
        break

# Plot the images in a 2x5 grid
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
axes = axes.ravel()  # Flatten axes array

for i in range(10):
    axes[i].imshow(sample_images[i], cmap='gray')
    axes[i].set_title(class_names[i])
    axes[i].axis('off')  # Hide axes

plt.tight_layout()
plt.show()
```
